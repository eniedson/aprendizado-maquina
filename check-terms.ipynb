#### Imports

```python
import os
import pandas as pd
```

#### Criando DataFrame com as colunas necessárias
Não é mais necessário executar o comando abaixo, uma vez que o arquivo já foi criado.

```python
columns = ['md_file', 'acordao', 'relatorio', 'voto']
df = pd.DataFrame(columns=columns)
df
```

#### Lendo o arquivo .csv

```python
df = pd.read_csv('./terms_verified.csv')
df
```

#### Acessando os arquivos, verificando se os termos desejados estão presentes e adicionando informações ao DataFrame

```python
directory_path = './md_files'

for filename in os.listdir(directory_path):
    file_path = os.path.join(directory_path, filename)
    
    if os.path.isfile(file_path):
        print(f'Acessando o arquivo: {file_path}')

        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            
            acordao_found = '# ACÓRDÃO' in content
            relatorio_found = '# RELATÓRIO' in content
            voto_found = '# VOTO' in content

            new_row = pd.DataFrame({
                'md_file': [file_path],
                'acordao': [acordao_found],
                'relatorio': [relatorio_found],
                'voto': [voto_found]
            })
            
            df = pd.concat([df, new_row], ignore_index=True)
```

```python
df.sample(10)
```

#### Salvando o DataFrame em formato .csv

```python
output_csv_path = './terms_verified.csv'
df.to_csv(output_csv_path, index=False, encoding='utf-8')
print(f'DataFrame salvo em {output_csv_path}')
```
